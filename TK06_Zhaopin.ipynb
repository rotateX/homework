```python
import requests
from random import randint
import csv
import time
from bs4 import BeautifulSoup


def get_desc(url):
    
    headers = {
    'Accept': 'pplication/json, text/plain, */*',
    'Origin': 'https://jobs.zhaopin.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'
    }
    cookies = {
        'x-zp-client-id': '1c7a3f60-8483-4869-8c48-08806fbf46bc'
    }
    desc_req = requests.get(url, headers=headers, cookies=cookies)
    desc_req.encoding = 'utf8'
    print(desc_req.url)
    time.sleep(3)
    desc_rep = desc_req.text
    soup = BeautifulSoup(desc_rep, 'html.parser')
#     print(soup)
    try:
        find_job_detail = soup.find('div', 'describtion')
        job_detail = soup.find('div', class_='describtion')
        print(job_detail)
        return job_detail.get_text(strip=True)
    except Exception as e:
        job_detail = 'None'
        print(e)
        return job_detail
        
with open('zhaopin.csv', 'w+', newline='', encoding='utf_8_sig') as f:
    csv_headers = [
    '岗位名称', 
    '薪水', 
    '学历要求', 
    '工作经验要求', 
    '招聘详细页url', 
    '是否全职',
    '公司名', 
    '公司类型', 
    '公司规模等', 
    '职位信息'
    ]
    f_csv = csv.writer(f)
    f_csv.writerow(csv_headers)
    for i in range(0, 90, 90):
        params = {
            'start': i,
            'pageSize': '90',
            'cityId': '682',
            'workExperience': '-1',
            'education': '-1',
            'companyType': '-1',
            'employmentType': '-1',
            'jobWelfareTag': '-1',
            'kw': 'python',
            'kt': '3'
        }
        url = 'https://fe-api.zhaopin.com/c/i/sou'
        try:
            req = requests.get(url, params=params)
            req.encoding = 'utf8'
            rep = req.json()
#             print(rep)
        except Exception as e:
            print(e)
        if req.status_code == 200 and rep['code'] == 200 :
            try:
                results = rep['data']['results']
                for job in results:
                    try:
                        jobName = job['jobName']  # 岗位名称
                        salary = job['salary']  # 薪水
                        eduLevel = job['eduLevel']['name']  # 学历要求
                        workingExp = job['workingExp']['name']  # 工作经验要求
                        positionURL = job['positionURL']  # 招聘详细页url
                        emplType = job['emplType']  # 是否全职
                        company_name = job['company']['name']  # 公司名
                        company_type = job['company']['type']['name']  # 公司类型
                        company_size = job['company']['size']['name']  # 公司规模等
                        job_detail = get_desc(positionURL)
                        # print(jobName, salary, eduLevel, workingExp, positionURL, emplType, company_name, company_type, company_size)
                        csv_row = [
                            jobName, 
                            salary, 
                            eduLevel, 
                            workingExp, 
                            positionURL, 
                            emplType, 
                            company_name, 
                            company_type, 
                            company_size,
                            job_detail
                        ]
                        f_csv.writerow(csv_row)
                    except Exception as nullerr:
                        print(nullerr)
                    time.sleep(randint(1, 4))
                print('finished！')
            except Exception as err:
                print(err)
        time.sleep(randint(1,4))
```
