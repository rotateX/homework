### 作业06：你的最好成绩

```python
# -*- coding : utf-8 -*-

import random

# 计算平均值
def avg(clist):
    sum_count = 0
    for i in clist:
        sum_count += i
    avg_count = sum_count/len(clist)
    return avg_count

def playsum(clist):
    sum_count = 0
    for i in clist:
        sum_count += i
    return sum_count

def saveinfo(playdata):
    with open('playrecord.txt', 'a+', encoding='utf-8') as f:            
        f.write(playdata)
        print('保存成功！')
    
        
def getplayinfo(pname):
    playinfo = {}
    with open('playrecord.txt', 'r+', encoding='utf-8') as f:
        for line in f:
            p_info = line.split(' ', 3)
            p_name = p_info[0].strip()
            p_rounds = p_info[1].strip()
            p_min = p_info[2].strip()
            p_count = p_info[3].strip()
            playinfo[p_name] = {'rounds': p_rounds, 'min': p_min, 'count': p_count}
    if playinfo.get(pname):
        info = ("玩家\'%s\'的历史信息: 总游戏回合%s，最快猜对次数%s，总次数%s。" % (
            pname, playinfo[p_name]['rounds'], playinfo[p_name]['min'], playinfo[p_name]['count'] )
             )
        
        print(info)
    else:
        info = ('欢迎新玩家！')
        print(info)
    return info
    

def playgame():
    play_round = 1 # 计算游戏回合
    count_list = [] # 每回合猜次数
    sign = True
    player = input('游戏开始，请输入玩家名：\n')
    getplayinfo(player)

    while sign is True:
        correct_num = random.randint(0, 100)
        print('==========第%d回合开始==========' % play_round)
        print('正确答案：%d'% correct_num)
        temp_count = 0  # temp每回合猜的次数
        while True:
            try:
                temp_count += 1
                guess_num = int(input('猜数字:\n'))
                if guess_num == correct_num:
                    print('你猜对惹！本回合共猜了%d次' % temp_count)
                    break
                elif guess_num > correct_num:
                    print('不对，请往小了猜。')

                elif guess_num < correct_num:
                    print('不对，请往大了猜。')

            except ValueError as verr:
                print('请输入数字,游戏重新开始！')
                temp_count = 0
                break
        count_list.append(temp_count)
        print('==========第%d回合结束==========' % play_round)
        play = input('再来一回合吗[yes/no]\n')
    #     print(count_list)
        if play == 'yes':
            play_round += 1
            sign = True
        else:
            min_count = min(count_list)
            avg_count = avg(count_list)
            sum_count = playsum(count_list)
            playdata = '%s %d %d %d \n' % (player, play_round, min_count, sum_count)
            saveinfo(playdata)
            print('游戏结束！')
            sign = False
            
if __name__ == '__main__':
    playgame()
```
