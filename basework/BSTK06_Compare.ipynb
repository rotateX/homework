### 作业06：你的最好成绩

```python
# -*- coding : utf-8 -*-

import random
import os

def playsum(clist):
    sum_count = 0
    for i in clist:
        sum_count += i
    return sum_count

def saveinfo(playdata):
    with open('playrecord.txt', 'w', encoding='utf-8') as f:
        f.write(playdata)
        print('保存成功！')
        
def getplayinfo(pname):
    """
    获取玩家历史信息，迭代playrecord.txt文件对象，
    每个对象通过split()分割，
    p_name：玩家名， p_round：游戏回合， p_min：最少次数， p_count：总次数
    """
    playinfo = {}
    try:
        with open('playrecord.txt', 'r+', encoding='utf-8') as f:
            print('打开成功')
            for line in f:
                p_info = line.split(' ', 3)
                p_name = p_info[0].strip()
                p_rounds = p_info[1].strip()
                p_min = p_info[2].strip()
                p_count = p_info[3].strip()
                # 玩家信息保存为字典，
                playinfo[p_name] = {'round': p_rounds, 'min': p_min, 'count': p_count}
    except FileNotFoundError:
        with open('playrecord.txt', 'w+', encoding='utf-8') as f:
            print('创建成功！')
            for line in f:
                p_info = line.split(' ', 3)
                p_name = p_info[0].strip()
                p_rounds = p_info[1].strip()
                p_min = p_info[2].strip()
                p_count = p_info[3].strip()
                # 玩家信息保存为字典，
                playinfo[p_name] = {'round': p_rounds, 'min': p_min, 'count': p_count}
    
    
    if playinfo.get(pname):
        # 玩家存在时，返回初始游戏回合、初始最少次数，初始总次数
        info = ("玩家\'%s\'的历史信息: 总游戏回合%s，最快猜对次数%s，总次数%s。" % (
            pname, playinfo[p_name]['round'], playinfo[p_name]['min'], playinfo[p_name]['count'] )
             )
        init_round, init_min, init_count = playinfo[p_name]['round'], playinfo[p_name]['min'], playinfo[p_name]['count']
    else:
        init_round, init_min, init_count = 0, 0, 0  # 玩家不存在时，初始化游戏回合0、最少次数1，总次数0
        info = ('欢迎新玩家！')
    return (info, init_round, init_min, init_count)
    

def playgame():
    play_round = 0 # 计算游戏回合
    count_list = [] # 每回合猜次数
    sign = True
    player = input('游戏开始，请输入玩家名：\n')
    info, init_round, init_min, init_count = getplayinfo(player)
    print(info)
    
    while sign is True:
        play_round += 1
        correct_num = random.randint(0, 100)
        print('==========第%d回合开始==========' % play_round)
        print('正确答案：%d'% correct_num)
        temp_count = 0  # temp每回合猜的次数
        while True:
            try:
                temp_count += 1
                guess_num = int(input('猜数字:\n'))
                if guess_num == correct_num:
                    print('你猜对惹！本回合共猜了%d次' % temp_count)
                    break
                elif guess_num > correct_num:
                    print('不对，请往小了猜。')
                elif guess_num < correct_num:
                    print('不对，请往大了猜。')

            except ValueError as verr:
                print('请输入数字,游戏重新开始！')
                temp_count = 0
                print('==========第%d回合结束==========' % play_round)
                play_round -= 1 # 输入错误则不计入本回合
                break
        count_list.append(temp_count)
        play = input('再来一回合吗[yes/no]\n')
#         print(count_list)
        if play == 'yes':
            sign = True
        else:
            temp_min_count = min(count_list)
#             print(temp_min_count)
            if int(init_min) == 0:
                min_count = temp_min_count
            elif temp_min_count > int(init_min):
                min_count = init_min
            else:
                min_count = temp_min_count
            sum_count = playsum(count_list) + int(init_count)
            sum_round = int(init_round) + play_round
            playdata = '%s %s %s %s \n' % (player, sum_round, min_count, sum_count)
            saveinfo(playdata)
            print('游戏结束！')
            sign = False
            
if __name__ == '__main__':
    playgame()
```
